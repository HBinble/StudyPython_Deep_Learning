```python
import numpy as np
```

```python
## 1. NAND w1, w2, theta
## 2. AND w1, w2, theta
## 3. OR w1, w2, theta
## NAND 논리를 출력하는 함수 
## AND 논리를 출력하는 함수 
## OR 논리를 출력하는 함수 
## XOR 논리를 출력하는 함수
```

```python
X = np.array([[0,0,1,1],[0,1,0,1]]).T
```

```python
X
```

```python
X[:,0]
```

```python
## 함수
def AND(x):
    W = np.array([[0.1],[0.1]])
    theta = 0.1
    return np.where(np.dot(x,W) > theta,1,0)

def OR(x):
    W = np.array([[0.2],[0.2]])
    theta = 0.1
    return np.where(np.dot(x,W) > theta,1,0)

def NAND(x):
    W = np.array([[0.1],[0.1]])
    theta = 0.2
    return np.where(np.dot(x,W) < theta,1,0)

def XOR(x):
    x = np.c_[NAND(x),OR(x)]
    return AND(x)
```

```python
print(AND(X))
print(OR(X))
print(NAND(X))
print(XOR(X))
```

```python
w1s = np.linspace(0.1,1,10)
w2s = np.linspace(0.1,1,10)
thetas = np.linspace(0.1,1,10)
for w1 in w1s:
    for w2 in w2s:
        for theta in thetas:
            result = np.where(X[:,0]*w1+X[:,1]*w2 > theta,1,0)
            if np.sum(result == np.array([0,0,0,1])) == 4:
                print(w1,w2,theta)
```
